```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
warnings.filterwarnings('ignore')
```

```python
df_cat = df.select_dtypes('object')
df_num = df.select_dtypes(include = ['int','float'])
```

```python
df_num.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df_cat:
    le = LabelEncoder()
    df_cat[col] = le.fit_transform(df_cat[col])
```

```python
df_cat
```

```python
df = pd.concat([df_cat,df_num],axis = 1)
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.25,random_state = 1)
```

```python
def build_models(model):
    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    print(confusion_matrix(ytest,ypred))
    print(classification_report(ytest,ypred))
    print('Training Score : ',model.score(xtrain,ytrain))
    print('Testing Score : ',model.score(xtest,ytest))
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC()
dt = DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier,VotingClassifier,AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
```

#### Bagging and RandomForest 

```python
def BuildBCRFModels(model):
    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    print(confusion_matrix(ytest,ypred))
    print(classification_report(ytest,ypred))
    print(f'Training Score: {model.score(xtrain,ytrain)}')
    print(f'Testing Score: {model.score(xtest,ytest)}')
```

```python
BClr = BaggingClassifier(LogisticRegression())
BCknn = BaggingClassifier(KNeighborsClassifier())
BCsvm = BaggingClassifier(SVC())
BCdt = BaggingClassifier(DecisionTreeClassifier())
rf = RandomForestClassifier()
```

```python
BuildBCRFModels(BClr)
```

```python
BuildBCRFModels(BCknn)
```

```python
BuildBCRFModels(BCsvm)
```

```python
BuildBCRFModels(BCdt)
```

```python
# Overfitting Scenario
BuildBCRFModels(rf)
```

#### VotingClassifier

```python
m1 = []
m1.append(('logistic regression',LogisticRegression()))
m1.append(('KNN',KNeighborsClassifier()))
m1.append(('SVM',SVC()))
m1.append(('decision tree',DecisionTreeClassifier()))
```

```python
vc = VotingClassifier(estimators = m1)
vc.fit(xtrain,ytrain)
ypred = vc.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
print(f'training score : {vc.score(xtrain,ytrain)}')
print(f'testing score : {vc.score(xtest,ytest)}')
```

#### Boosting

```python
def Buildboostmodels(models):
    models.fit(xtrain,ytrain)
    ypred = models.predict(xtest)
    print(confusion_matrix(ytest,ypred))
    print(classification_report(ytest,ypred))
    print(f'Training Score: {models.score(xtrain,ytrain)}')
    print(f'Testing Score: {models.score(xtest,ytest)}')
```

```python
ada = AdaBoostClassifier()
gd = GradientBoostingClassifier()
xgd = XGBClassifier()
```
